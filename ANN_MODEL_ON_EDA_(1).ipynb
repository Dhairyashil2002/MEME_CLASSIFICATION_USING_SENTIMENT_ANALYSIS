```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader, TensorDataset
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from torch.cuda.amp import autocast, GradScaler
```

```python
df = pd.read_csv('labels.csv')

# Keep relevant columns
df = df[['image_name', 'text_ocr', 'overall_sentiment']]

# Label Encoding for 'overall_sentiment'
label_encoder = LabelEncoder()
df['encoded_sentiment'] = label_encoder.fit_transform(df['overall_sentiment'])
```

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Tokenization and Data Formatting
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
train_encodings = tokenizer.batch_encode_plus(
    train_df['text_ocr'].astype(str).tolist(),
    add_special_tokens=True,
    return_attention_mask=True,
    truncation=True,  # Explicitly activate truncation
    padding='max_length',  # Pad to max_length
    max_length=256,
    return_tensors='pt'
)
```

```python
test_encodings = tokenizer.batch_encode_plus(
    test_df['text_ocr'].astype(str).tolist(),
    add_special_tokens=True,
    return_attention_mask=True,
    truncation=True,
    padding='max_length',
    max_length=256,
    return_tensors='pt'
)
```

```python
train_dataset = TensorDataset(
    train_encodings['input_ids'],
    train_encodings['attention_mask'],
    torch.tensor(train_df['encoded_sentiment'].astype(int).tolist())
)
```

```python
test_dataset = TensorDataset(
    test_encodings['input_ids'],
    test_encodings['attention_mask'],
    torch.tensor(test_df['encoded_sentiment'].astype(int).tolist())
)
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Model Definition and Training
NUM_CLASSES = len(df['encoded_sentiment'].unique())
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=NUM_CLASSES)
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)



```

```python
# Training loop with gradient accumulation and mixed-precision training
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
scaler = GradScaler()

accumulation_steps = 4  # Accumulate gradients over 4 batches before updating
num_epochs = 3

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for i, batch in enumerate(train_loader):
        inputs, attention_mask, labels = batch
        inputs, attention_mask, labels = inputs.to(device), attention_mask.to(device), labels.to(device)

        optimizer.zero_grad()

        with autocast():
            outputs = model(inputs, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss

        scaler.scale(loss).backward()

        if (i + 1) % accumulation_steps == 0:
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()

        running_loss += loss.item()

        if i % 100 == 99:  # Print every 100 batches
            print(f'Epoch {epoch + 1}, Batch {i + 1}, Loss: {running_loss / 100:.4f}')
            running_loss = 0.0
```

```python
# ... (previous code)

# Evaluation
model.eval()
with torch.no_grad():
    correct = 0
    total = 0
    test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)  # Reducing batch size for evaluation

    for batch in test_loader:
        inputs, attention_mask, labels = batch
        inputs, attention_mask, labels = inputs.to(device), attention_mask.to(device), labels.to(device)

        outputs = model(inputs, attention_mask=attention_mask)
        _, predicted = torch.max(outputs.logits, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = correct / total
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
import pandas as pd
df = pd.read_csv('labels.csv')

# Keep relevant columns
df = df[['image_name', 'text_ocr', 'overall_sentiment']]
```

```python
from sklearn.preprocessing import LabelEncoder
from transformers import BertTokenizer

# Extract relevant columns
text_data = df['text_ocr'].astype(str).tolist()
label_data = df['overall_sentiment'].astype(str).tolist()
image_name_data = df['image_name'].astype(str).tolist()

# Encode string labels into numerical values
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(label_data)

# Tokenization and encoding
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
encoded_inputs = tokenizer(text_data, padding=True, truncation=True, return_tensors='pt')
```

```python
# Create datasets and dataloaders
class YourDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

# Use encoded_labels instead of label_data
labels_dataset = YourDataset(encoded_inputs, encoded_labels)
```

```python
# Split the dataset into train, validation, and test sets
train_size = int(0.8 * len(labels_dataset))
val_size = int(0.1 * len(labels_dataset))
test_size = len(labels_dataset) - train_size - val_size

train_dataset, val_dataset, test_dataset = torch.utils.data.random_split(labels_dataset, [train_size, val_size, test_size])

# Create DataLoader instances
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
# Cell 4: Model Definition and Training Setup
NUM_CLASSES = len(set(label_data))  # Assuming train_labels is a list or array
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=NUM_CLASSES)
model.bert.encoder.layer[-1].output.dropout = torch.nn.Dropout(0.1)
model.bert.pooler.dropout = torch.nn.Dropout(0.1)
model.classifier.dropout = torch.nn.Dropout(0.1)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
optimizer = AdamW(model.parameters(), lr=2e-5)
scaler = GradScaler()
accumulation_steps = 4
num_epochs = 5
num_fine_tune_epochs = 3
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=len(train_loader) * num_epochs)
```

```python
# Cell 5.1: Training Loop Setup
best_val_accuracy = 0.0
early_stopping_counter = 0
patience = 3
```

```python
# Cell 5.2: Training Loop
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for i, batch in enumerate(tqdm(train_loader, desc=f'Epoch {epoch + 1}/{num_epochs}')):
        inputs = {key: val.to(device) for key, val in batch.items() if key != 'labels'}
        labels = batch['labels'].to(device)

        optimizer.zero_grad()

        with autocast():
            outputs = model(**inputs, labels=labels)
            loss = outputs.loss

        scaler.scale(loss).backward()

        if (i + 1) % accumulation_steps == 0:
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()

        running_loss += loss.item()
        scheduler.step()
```
